```python
# Function to calculate grade based on percentage
def calculate_grade(percentage):
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B'
    elif percentage >= 60:
        return 'C'
    elif percentage >= 50:
        return 'D'
    else:
        return 'F'

# Function to get student data
def get_student_data():
    total_student_data = []  # List to store all student records

    while True:
        single_student_data = {}
        single_student_data['name'] = input("Enter Your Name: ")
        single_student_data['roll number'] = input("Enter Your Roll Number: ")
        single_student_data['marks'] = {}
        subjects = ["English", "Urdu", "Maths", "Physics", "Chemistry", "Computer"]

        for subject in subjects:
            while True:
                try:
                    marks = int(input(f"Enter Your Marks in {subject}: "))
                    if 0 <= marks <= 100:
                        single_student_data['marks'][subject] = marks
                        break  # Exit loop if valid marks are entered
                    else:
                        print("Invalid marks. Marks should be between 0 to 100! Please enter again.")
                except ValueError:
                    print("Invalid input. Please enter a valid number.")

        total_student_data.append(single_student_data)  # Append only after collecting all subject marks
        print(f"Record of {single_student_data['name']} inserted successfully.\n")

        more = input("Do you want to insert more students? Press 'Y' for Yes or 'N' for No: ").strip().lower()
        while more not in ['y', 'n']:
            more = input("Invalid choice! Press 'Y' for Yes or 'N' for No: ").strip().lower()
        if more == 'n':
            break

    return total_student_data  # Return the list of all students

# Function to generate report based on student data
def generate_report(students):
    print("\nStudent Report Cards")
    print("=" * 40)

    for student in students:
        print(f"Student Name: {student['name']}")
        print(f"Roll Number: {student['roll number']}")
        print("Subject-wise Marks:")
        total_marks = sum(student['marks'].values())

        for subject, marks in student['marks'].items():
            print(f"  {subject}: {marks}")

        percentage = total_marks / 6  # 6 subjects
        grade = calculate_grade(percentage)

        print(f"Total Marks: {total_marks}/600")
        print(f"Percentage: {percentage:.2f}%")
        print(f"Grade: {grade}")
        print("=" * 40)

# Function to save report to a text file
def save_report_to_file(students):
    with open("student_report.txt", "w") as file:
        file.write("Student Report Cards\n")
        file.write("=" * 40 + "\n")

        for student in students:
            file.write(f"Student Name: {student['name']}\n")
            file.write(f"Roll Number: {student['roll number']}\n")
            file.write("Subject-wise Marks:\n")

            total_marks = sum(student['marks'].values())
            for subject, marks in student['marks'].items():
                file.write(f"  {subject}: {marks}\n")

            percentage = total_marks / 6  # 6 subjects
            grade = calculate_grade(percentage)

            file.write(f"Total Marks: {total_marks}/600\n")
            file.write(f"Percentage: {percentage:.2f}%\n")
            file.write(f"Grade: {grade}\n")
            file.write("=" * 40 + "\n")

    print("Report saved to student_report.txt successfully!")

# Main function to run the program
def main():
    print("Welcome to the Student Report Card Generator")
    students = get_student_data()  # Collect student data
    generate_report(students)  # Display report
    save_report_to_file(students)  # Save report to file

if __name__ == "__main__":
    main()
```
